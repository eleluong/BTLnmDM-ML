# Clean data và tính toán các trường cần tính

```python
df = data.load_data()
df.head(10)
```

```python
df = df[['ticker', 'date', 'open', 'close']]
df.head()
```

Tính log return

```python
df['log return'] = np.log(df['close']/df['close'].shift(1))
df.fillna(0, inplace = True)
```

Tính volatility

```python
df['volatility'] = df['log return'].rolling(30).std() *np.sqrt(252)
```
